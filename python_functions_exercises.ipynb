Python Functions Exercises

```python
# 1
def greet():
    print('Hello, world!')

greet()
```

```python
# 2
def show_name(name):
    print('Hello', name)

show_name('Ximena')
```

```python
# 3
def add(a, b):
    return a + b

print(add(3, 5))
```

```python
# 4
def subtract(a, b):
    return a - b

print(subtract(10, 4))
```

```python
# 5
def multiply(a, b):
    return a * b

print(multiply(3, 6))
```

```python
# 6
def divide(a, b):
    return a / b

print(divide(10, 2))
```

```python
# 7
def is_even(number):
    return number % 2 == 0

print(is_even(4))
```

```python
# 8
def find_max(a, b):
    return a if a > b else b

print(find_max(7, 10))
```

```python
# 9
def square(number):
    return number ** 2

print(square(5))
```

```python
# 10
def cube(number):
    return number ** 3

print(cube(3))
```

```python
# 11
def count_letters(word):
    return len(word)

print(count_letters('python'))
```

```python
# 12
def convert_to_upper(text):
    return text.upper()

print(convert_to_upper('hello'))
```

```python
# 13
def calculate_area(length, width):
    return length * width

print(calculate_area(5, 4))
```

```python
# 14
def average(numbers):
    return sum(numbers) / len(numbers)

print(average([1,2,3,4,5]))
```

```python
# 15
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

print(factorial(5))
```
